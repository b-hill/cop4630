### Imports

```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from math import e
%matplotlib inline
```

#Graphs

## Function 1


###Definition

```
def f1(x, y):
    return (x**2+y**2)
```

###3D

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.text(0,0,10, "P1", color="red", zorder=1000, fontsize=16)
ax.plot([0.], [0.], [0.],  markerfacecolor='r', markeredgecolor='r', marker='o', markersize=10, alpha=0.9, zorder=100)
```

###Contour

```
plt.contour(X, Y, Z, colors='black');
plt.text(0.5,0.5, "P1", color="red", zorder=1000, fontsize=16)
plt.plot([0.], [0.], markerfacecolor='r', markeredgecolor='r', marker='o', markersize=10, alpha=0.9, zorder=10000)
```

###Density

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.text(0.5,0.5, "P1", color="white", zorder=1000, fontsize=16)
plt.plot([0.], [0.], markerfacecolor='w', markeredgecolor='w', marker='o', markersize=10, alpha=0.9, zorder=10000)
```
